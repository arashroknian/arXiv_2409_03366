```python
def riga(ks, axs, k=None):
    if k is None: k = ks.l
    mv = np.array( ks.hist['mv_count'] )
    theta = np.stack( ks.hist['theta'] )
    eps = np.stack( ks.hist['eps'] )
    # for j in range(10): 
    for j in range(k): 
        axs[0].plot(mv, theta[:,j], linewidth=0.5)
        axs[1].plot(mv, eps[:,j], linewidth=0.5)
    axs[0].set_yscale('symlog')
    axs[1].set_yscale('log')
```

```python
def arpack(pb, k, tol, m, vals_c):
    A = pb.eig_ops().S
    rng1 = np.random.default_rng(2023); v0 = rng1.random(A.shape[0])
    pb._matvec_count = 0
    try: vals,_ = sps.linalg.eigs(A, k=k, which='LR', v0=v0, tol=tol, ncv=m)
    except: print('no convergence')
    eps = 1 - vals/vals_c
    return pb._matvec_count, vals, eps

def ks(pb, k, tol, m, vals_c):
    A = pb.eig_ops().S
    b = pb._M.diagonal()
    ks = DKS(A, k, m, b=b, tol=tol)
    pb._matvec_count = 0
    ks.itera(maxmv=10000, log_freq=-1)
    assert pb._matvec_count == ks.mv_count
    vals = ks.eig()[0][:k]
    eps = 1 - vals/vals_c
    return ks.mv_count, vals, eps
```

```python
pb = co.pkl(dict(cartella='../simulazioni', nome='eig_vg14_D1g')); A = pb.eig_ops().S; b = pb._M.diagonal()
vals_c,_ = sps.linalg.eigs(A, k=1, which='LR')
```

```python
print('')
k, tol, m = 1, 1e-2, 32
%time print( arpack(pb, k, tol, m, vals_c) )
%time print( ks(pb, k, tol, m, vals_c) )

print('')
k, tol, m = 1, 1e-2, 64
%time print( arpack(pb, k, tol, m, vals_c) )
%time print( ks(pb, k, tol, m, vals_c) )

print('')
k, tol, m = 1, 1e-2, 128
%time print( arpack(pb, k, tol, m, vals_c) )
%time print( ks(pb, k, tol, m, vals_c) )
```
